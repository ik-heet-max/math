### Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math
```

```python
k_1 = 1
k_2 = 7
```

```python
x = np.linspace(0, 2*math.pi, 400)
a = np.cos(k_1*x)
b = np.cos(k_2*x)
```

```python
plt.plot(x, a, 'r')
plt.plot(x, b, 'g')
```

### Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.


```python
x = np.linspace(0, math.pi)
a = np.sin(x)
b = np.cos(x)
plt.xlim(-2, 2)
plt.ylim(-1.3, 1.3)
plt.plot(a, b, 'b')
plt.plot(-a, b, 'b')
```

```python
x = np.linspace(0, math.pi)
a = np.sin(x)
b = np.cos(x)
plt.xlim(-2, 2)
plt.ylim(-1.3, 1.3)
plt.plot(1.5*a, 0.75*b, 'b')
plt.plot(-1.5*a, 0.75*b, 'b')
```

### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def length(x, y, z):
    return np.linalg.norm([x, y, z])
```

### Нарисуйте трехмерный график двух параллельных плоскостей

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = x + y + 4
z1 = x + y + 7

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.jet)
ax.plot_surface(x, y, z1, cmap=plt.cm.jet)
```

### Изобразим параболоид

```python
x = np.outer(np.ones(30,), np.linspace(-4, 4, 30))    
y = x.copy().T

z = x ** 2 + y ** 2

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.jet)
```

### Изобразим эллипсоид

```python
x = np.outer(np.ones(360,), np.linspace(-360, math.pi, 360))    
y = x.copy().T

a = np.cos(x) * np.cos(y)
b = np.cos(x) * np.sin(y)
c = np.sin(x)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(a, b, c)
```
